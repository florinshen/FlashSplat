```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import cv2

def show_points(coords, labels, ax, marker_size=375):
    pos_points = coords[labels==1]
    neg_points = coords[labels==0]
    ax.scatter(pos_points[:, 0], pos_points[:, 1], color='green', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)
    ax.scatter(neg_points[:, 0], neg_points[:, 1], color='red', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)   

# image_path = "/home/shenqiuhong/Downloads/Gaussian-Grouping/data/lerf_mask/teatime/images/frame_00001.jpg"

# image_path = "/home/shenqiuhong/Downloads/tandt/tandt/truck/images/000001.jpg"
# image_path = "/home/shenqiuhong/Downloads/Gaussian-Grouping/data/mipnerf360/counter/images_2/DSCF5861.JPG"
image_path = "/home/shenqiuhong/Downloads/llff-gs/horns/images_4/DJI_20200223_163016_842.png"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Pixel position to visualize (x, y)
plt.figure(figsize=(10,10))
plt.imshow(image)
# (500, 500)
# for truck, (200, 280), (700, 300)
input_point = np.array([[550, 500], [300, 300], [500, 200], [600, 300], [420, 250], [600, 240], [620, 600], [630, 420]])
input_label = np.array([1] * input_point.shape[0])
show_points(input_point, input_label, plt.gca())
plt.axis('on')
plt.show()  
```
